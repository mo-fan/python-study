## 4.4.2 网络爬虫

```python
import requests
from lxml import etree
```

```python
header = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36'}
url = 'https://www.kugou.com/yy/rank/home/1-8888.html?from=homepage'
```

```python
r = requests.get(url=url,headers=header)
page_text = r.text
r.encoding = r.apparent_encoding
page_text
```

```python
#song_list = tree.xpath("//div[@id='rankWrap']/div[2]/ul")
tree = etree.HTML(page_text)
song_list = tree.xpath("//*[@id='rankWrap']/div[2]/ul/li")
song_list
```

```python
data=[]
for li in song_list:
    rank = tree.xpath('./span[3]//text()')
    singer = ''.join(li.xpath('./a/text()')).split('-')[0]
    song = ''.join(li.xpath('./a/text()')).split('-')[1]
    time = tree.xpath('./span[4]/span/text()')
    a={'rank':rank,
      'singer':singer,
      'song':song,
      'time':time}
    data.append(a) 
data
    
```

```python
from pandas import DataFrame
```

```python
df = DataFrame(data)
df
```

```python
soup = BeautifulSoup(page_text,'lxml')
```

```python

data=[]

ranks = soup.select('span.pc_temp_num')
times = soup.select('span.pc_temp_tips_r>span')
titles = soup.select('div.pc_temp_songlist>ul>li>a')
for rank,time,title in zip(ranks,times,titles):
    rank = rank.get_text().strip()
    time = time.get_text().strip()
    singer = title.get_text().split('-')[0]
    song = title.get_text().split('-')[1]
    a={'rank':rank,
      'time':time,
      'singer':singer,
       'song':song}
    data.append(a) 
data
```

```python
df = DataFrame(data)
df
```
